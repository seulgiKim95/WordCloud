```python
df = pd.read_excel('../output/Jobkorea_0725_1.xlsx', index_col=0)
df.head()
```

```python
remove_rows = []
for i in df.index:
    if df['cname'][i] == 'None':
        remove_rows.append(i)
df.drop(remove_rows, inplace=True)
```
