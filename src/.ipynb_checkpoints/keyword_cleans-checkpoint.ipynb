```python
import pandas as pd
```

```python
def keyword_cleans(x):
    # x에 키워드 열을 넣으면 됩니다.
    # jobs_list['keyword']=keyword_cleans(jobs_list['keyword'])

    import numpy as np
    import re
    
    x=x.str.split(', ')
    # 각 단어로 구분하였습니다.
    # 여긴 ', '로 모든 단어들을 나눌 수 있었지만 다른 사이트엔 다르게
    # 적용할 필요성이 있어 보입니다.
    
    p=re.compile('채용관|일자리|총무|프로그래머|엔지니어|디자이너|개발자|관리자|\
                교육|테스터|게임|시스템분석|강사|사무|법무|컨텐츠|교육|제휴|의료|\
                경영|전략|인사|노무|기획|투자|은행|광고|고객|여행|영업|보험|보건|\
                쇼핑|오픈마켓|마케팅|자산|신문|잡지|언론|경영분석|컨설턴트|영양')
    # 우선 ~채용관, ~일자리, 을 삭제후, 직무, 저희가 원하는 스킬과 관련없는
    # 단어들을 삭제했습니다.
    
    for i in range(len(x)):
        d_list=['']
        for j in x[i]:
            t=p.search(j)
            if t!=None:
                d_list.append(j)
        x[i]=[elem for elem in x[i] if elem not in d_list]
     # 삭제할 요소들을 새 리스트에 따로 모아두고, 한 번에 제거했습니다.
    
    return x

# 이상하게 쇼핑을 제거리스트에 넣었는데도 쇼핑몰은 제거되지가 않았습니다..
# 혹시 시간되시면 확인가능하신가요?
# 프로그래머를 넣을시 웹프로그래머나 응용프로그래머는 모두 삭제되었습니다.
```

```python
jobs_list=pd.read_excel('../data/Jobkorea_0725.xlsx')
jobs_list['keyword']=keyword_cleans(jobs_list['keyword'])
```
