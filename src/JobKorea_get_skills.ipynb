```python
df = pd.read_excel('../data/Jobkorea_0725.xlsx', index_col=0)
```

```python
# skills = df.skills[df.skills.notnull()]
tmp_job = df[df['job'] == '데이터 엔지니어']
skills = tmp_job.skills[tmp_job.skills.notnull()]
```

```python
import copy
df1 = copy.copy(df)
skills = df1.skills[df1.skills.notnull()]
drop_idx = []
for idx in skills.index:
    skill = skills[idx]
    if skill.find('<span ') >= 0:
        drop_idx.append(idx)
#     if skill.count('명') > 1:
#         print (idx)
#         print (skill)
df1.drop(drop_idx, inplace=True)
skills = df1.skills[df1.skills.notnull()]
skills.index
```

```python
sk = ' '.join(skills)
remove = ['\n', '/']
for rem in remove:
    sk = sk.replace(rem, '')
```

```python
okt = Okt()
# tokens = okt.pos(sk, norm=True, stem=True)
def tokenize(sentence):
    return ['/'.join(t) for t in okt.pos(sentence, norm=True, stem=True)]
skills_pos = tokenize(sk)
skills_pos
# tokens = [t for d in train_docs for t in d[0]]
# tokens
# tokens = okt.nouns(sk)
```

```python
# English
import nltk

# tokenize_en = nltk.tokenize.WordPunctTokenizer() # 띄어쓰기 단위 토크나이저
tokenize_en = nltk.tokenize.TreebankWordTokenizer()
en = [pos.split('/')[0] for pos in skills_pos if 'Alpha' in pos]
en = ' '.join(en)
binary_word = {'Big Data':'Big-Data', 'Big Query':'BigQuery',
               'Extract Transform Load':'ETL', 'data':'Data',
               'javascript':'JS', 'MYSQL':'MySQL'}
for bi_word, res in binary_word.items():
    en = en.replace(bi_word, res)
```

```python
try:
    en = tokenize_en.tokenize(en)
except:
    en = tokenize_en.tokenize(' '.join(en))
en = nltk.Text(en)
# en.collocations()

from wordcloud import STOPWORDS


stopword_en = set(STOPWORDS)
stopword_en.update(['font', 'family', 'mso', 'black', 'letter', 'color', 'spacing',
               'bidi', 'ascii', 'fareast', 'hansi', 'span', 'lang',
               'kerning', 'size', 'style', 'pt', 'On', 'gt', 'EN', 'US', 'AS'])
en = [each_word for each_word in en if each_word not in stopword_en]
en = nltk.Text(en)
en.collocation_list()
# en = ' '.join(en)
# en.find('Extract Transform Load')
```

```python
from wordcloud import WordCloud
data_en = en.vocab().most_common(300)

# wc = WordCloud(background_color='black',
#                max_words=2000,
#                mask=mask,
#                stopwords=stopwords
#               ).generate(data_en)
wc = WordCloud(background_color='white',
              relative_scaling=0.1).generate_from_frequencies(dict(data_en))
plt.figure(figsize=(16,8))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.savefig('../output/en.png')
# print (data_en)
# en.concordance('premise')
```
