```python
import copy
df1 = copy.copy(df)
skills = df1.skills[df1.skills.notnull()]
drop_idx = []
for idx in skills.index:
    skill = skills[idx]
    if skill.find('<span ') >= 0:
        drop_idx.append(idx)
#     if skill.count('명') > 1:
#         print (idx)
#         print (skill)
df1.drop(drop_idx, inplace=True)
skills = df1.skills[df1.skills.notnull()]
skills.index
```

```python
sk = ' '.join(skills)
remove = ['\n', '/']
for rem in remove:
    sk = sk.replace(rem, '')
```

```python
# tokens = okt.pos(sk, norm=True, stem=True)
def tokenize(sentence):
    return ['/'.join(t) for t in okt.pos(sentence, norm=True, stem=True)]
skills_pos = tokenize(sk)
skills_pos
# tokens = [t for d in train_docs for t in d[0]]
# tokens
# tokens = okt.nouns(sk)
```

```python
# English

# tokenize_en = nltk.tokenize.WordPunctTokenizer() # 띄어쓰기 단위 토크나이저
tokenize_en = nltk.tokenize.TreebankWordTokenizer()
en = [pos.split('/')[0] for pos in skills_pos if 'Alpha' in pos]
en = ' '.join(en)
binary_word = {'Big Data':'Big-Data', 'Big Query':'BigQuery',
               'Extract Transform Load':'ETL', 'data':'Data',
               'javascript':'JS'}
for bi_word, res in binary_word.items():
    en = en.replace(bi_word, res)
```

```python
try:
    en = tokenize_en.tokenize(en)
except:
    en = tokenize_en.tokenize(' '.join(en))
en = nltk.Text(en)
# en.collocations()
from wordcloud import STOPWORDS

stopword_en = set(STOPWORDS)
stopword_en.update(['font', 'family', 'mso', 'black', 'letter', 'color', 'spacing',
               'bidi', 'ascii', 'fareast', 'hansi', 'span', 'lang',
               'kerning', 'size', 'style', 'pt', 'On', 'gt', 'EN', 'US', 'AS'])
en = [each_word for each_word in en if each_word not in stopword_en]
en = nltk.Text(en)
en.collocations()
# en = ' '.join(en)
# en.find('Extract Transform Load')
```
